```python
import turtle

# A function to draw a rectangle and fills it with the color
def draw_rectangle(length,height,color):
    # set the fillcolor
    turtle.fillcolor(color) 
    
    # start the filling color
    turtle.begin_fill() 
    
    # drawing the rectangle
    for i in range(2):
        turtle.forward(length)
        turtle.right(90)
        turtle.forward(height)
        turtle.right(90)
    
    # ending the filling of the color        
    turtle.end_fill() 

# A function to draw a star and fills it with the color
def draw_star(size,color):
    # set the fillcolor
    turtle.fillcolor(color)

    # start the filling color
    turtle.begin_fill()
  
    # drawing the star
    for i in range(5):
        turtle.penup()
        turtle.forward(size)
        turtle.right(144)
  
    # ending the filling of color
    turtle.end_fill()
        
# A function to determine rgb colors from a parameter color string
def get_color(color):
    if color == "red":
        return 1,0,0
    if color == "white":
        return 1,1,1
    if color == "blue":
        return 0,0,1
    else:
        return 0,0,0

# A function draw_flag to draw a flag of a certain height
def draw_flag(a): 
    # Hoist (height) of the flag: A = 1.0
    # Fly (width) of the flag: B = 1.9
    b = 1.9 * float(a) 
    # Hoist (height) of the canton ("union"): C = 0.5385 (A × 7/13, spanning seven stripes)
    c = a * (7/13)
    # Fly (width) of the canton: D = 0.76 (B × 2/5, two-fifths of the flag width)
    d = b * (2/5)
    # Width of stripe: L = 0.0769 (A/13, One thirteenth of the flag height)
    l = a / 13
    # Diameter of star: K = 0.0616 (L × 4/5, four-fifths of the stripe width)
    k = l * (4/5)
    # Vertical distance between stars: C/10, One-tenth of the height of the canton)
    e = c / 10
    # Horizontal distance between stars: D/12, One twelfth of the width of the canton)
    g = d / 12
    
    # Draw American flag outline
    x = -150
    y = 150
    turtle.penup()
    turtle.goto(x,y)
    turtle.pendown()
    draw_rectangle(b,a,"white")
    
    # Draw 7 red stripes of flag
    for i in range(7):
        turtle.penup()
        turtle.goto(x,y)
        turtle.pendown()
        draw_rectangle(b,l,"red")
        y = y - l * 2
    
    # Draw canton
    turtle.penup()
    turtle.goto(-150,150)
    turtle.pendown()
    draw_rectangle(d,c,"blue")
    
    # Draw stars, setting up coordinates first
    x1 = -150 + g/2
    y1 = 150 - e/1.2
    
    x2 = -150 + g*1.5
    y2 = 150 - e*1.8
    
    # Draw 5 rows of 6 stars:
    for i in range(5):
        for j in range(6):
            turtle.penup()
            turtle.goto(x1,y1)
            turtle.pendown()
            draw_star(k,"white")
            x1 = x1 + g*2
        x1 = -150 + g/2
        y1 = y1 - e*2
    
    # Draw 4 rows of 5 stars:
    for i in range(4):
        for j in range(5):
            turtle.penup()
            turtle.goto(x2,y2)
            turtle.pendown()
            draw_star(k,"white")
            x2 = x2 + g*2
        x2 = -150 + g*1.5
        y2 = y2 - e*2

# The main function that calls the four functions above. 
def main():
    turtle.penup()
    turtle.goto(-400,250)
    turtle.pendown()
    draw_rectangle(200,100,"green")
    draw_star(50,"yellow")
    print(get_color("red"))
    draw_flag(400)
    
if __name__ == "__main__":
    main()
```
