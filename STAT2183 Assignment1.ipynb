```r
#import data
data <- read.csv(file = "C:/Users/chane/desktop/covid.csv") 
```

```r
# Group by regions using library dplyr
Grouped_data <- data %>% group_by(Region) %>% group_split()
```

```r
Midwest_death <- Grouped_data[[1]]$Deaths
#Check Normality
shapiro.test(Midwest_death)
qqnorm(Midwest_death)
qqline(Midwest_death)
       
#Normal - t.test       
t.test(Midwest_death, mu = 300)
hist(Midwest_death)
```

```r
Northeast_death <- Grouped_data[[2]]$Deaths
shapiro.test(Northeast_death)
qqnorm(Northeast_death)
qqline(Northeast_death)

#Normal t.test
t.test(Northeast_death, mu = 300)
hist(Northeast_death)
```

```r
South_death <- Grouped_data[[3]]$Deaths
shapiro.test(South_death)
qqnorm(South_death)
qqline(South_death)

#Not normally distributed
#Sign test for median
library(signmedian.test)

signmedian.test(South_death, mu = 300)
hist(South_death)
```

```r
West_death <- Grouped_data[[4]]$Deaths
shapiro.test(West_death)
qqnorm(West_death)
qqline(West_death)

t.test(West_death, mu = 300)
hist(West_death)
```
