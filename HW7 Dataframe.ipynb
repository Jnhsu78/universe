```r
#Q1
Data_Frame <- data.frame (
  id = c(1,2,3,4,5),
  name = c("Peter","Amy","Ryan","Gary","Michelle"),
  salary = c(623.30,515.20,611.00,729.00,843.25)
)

Data_Frame
```

```r
New_col_DF <- cbind(Data_Frame, department = c( "IT","math","finance","history","art"))

New_col_DF
```

```r
Data_Frame_New <- Data_Frame[c(1,3,5),c(2,3)]

Data_Frame_New
```

```r
# x-axis values
x <- c("Peter","Gary","Michelle")

# y-axis values
y <- c(Data_Frame$"salary"[1],Data_Frame$"salary"[4],Data_Frame$"salary"[5])

barplot(y, names.arg = x)
```

```r
# Create a vector of labels
mylabel <- c("highest salary", "lowest salary", "median salary")

# Create a vector of colors
colors <- c("blue", "yellow","green")

x <- c(min(Data_Frame$"salary"),min(Data_Frame$"salary"),median(Data_Frame$"salary"))

# Display the pie chart with colors
pie(x, label = mylabel, main = "Pie Chart", col = colors)

# Display the explanation box
legend("bottomleft", mylabel, fill = colors)
```
