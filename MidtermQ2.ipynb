```python
import pandas as pd

df = pd.read_csv('C:/Users/chane/Desktop/amazon-orders.csv')
df.head()
```

```python
df.shape
```

```python
df = df.fillna(value=0)
df.head()
```

```python
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
data = {
    "Sum":{"Amount($)":df["Total Charged"].sum()},
    "Mean":{"Amount($)":df["Total Charged"].mean()},
    "Median":{"Amount($)":df["Total Charged"].median()},
    "Max":{"Amount($)":df["Total Charged"].max()},
    "Min":{"Amount($)":df["Total Charged"].min()}
}

obj = pd.DataFrame(data)
print(obj)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
%matplotlib inline
```

```python
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
monthly_orders = df.groupby(pd.Grouper(key='Order Date', axis=0, 
                      freq='M')).sum()["Total Charged"]
monthly_orders.head()
```

```python
monthly_orders.plot.bar(ylabel='Total Charged',figsize=(20,10))
```
